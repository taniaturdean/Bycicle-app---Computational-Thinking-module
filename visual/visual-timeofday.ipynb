```python
import pandas as pd


engdata = pd.read_excel("allseverities.xlsx")
```

```python
# Importing Matplotlib
import matplotlib.pyplot as plt

# Ensure that graphs are displayed within the notebook
%matplotlib inline
```

```python
x=engdata["Time"]
y1=engdata["2011-"]
y2=engdata["2012-"]
y3=engdata["2013-"]
```

```python
# Changing back to default style
plt.style.use('default')

 
# create a color palette
palette = plt.get_cmap('Set3')

# Produces the figure
plt.figure(figsize=(14,8))

num=-1
for column in engdata.drop('Time', axis=1):
    num+=1
    plt.plot( x,engdata[column], color=palette(num), linewidth=1, alpha=0.9, label=column,marker='o')


# Axes labels
plt.xlabel("Time of the day")
plt.ylabel("All severities from accidents")

# Add legend
plt.legend(loc=2, ncol=2)



# Adds a title
plt.title("UK number of road accidents casualties - 2003-2013")

plt.savefig("figure112.png") # save as png
```
